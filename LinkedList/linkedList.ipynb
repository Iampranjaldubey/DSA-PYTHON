```python
class Node:
    def __init__(self, data):
        self.data = data  # Store the data
        self.next = None  # Pointer to the next node
        self.prev = None  # Pointer to the previous node


class doublyLinkendlist:
    def __init__(self):
        self.head = None  # First node (head) of the list
        self.tail = None  # Last node (tail) of the list

    def insert_at_end(self, data):
        """Insert a new node at the end of the list."""
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node  # Link the last node to the new node
            new_node.prev = self.tail  # Link the new node to the previous node
            self.tail = new_node  # Update the tail

    def insert_at_beginning(self, data):
        """Insert a new node at the beginning of the list."""
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def display_forward(self):
        """Display the list from head to tail."""
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next
        print("None")

    def display_backward(self):
        """Display the list from tail to head."""
        temp = self.tail
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.prev
        print("None")


# Create a doubly linked list object
dll = doublyLinkendlist()

# Insert elements at the end
dll.insert_at_end(10)
dll.insert_at_end(20)
dll.insert_at_end(30)

# Insert element at the beginning
dll.insert_at_beginning(5)

print("Forward display:")
dll.display_forward()

print("Backward display:")
dll.display_backward()

         


```

```python
class Node:
    """Class to represent a node in a circular linked list."""
    def __init__(self, data):
        self.data = data  # Store node value
        self.next = None  # Pointer to next node


class CircularLinkedList:
    """Class to implement a circular linked list."""
    
    def __init__(self):
        self.head = None  # Initialize an empty circular linked list
        self.size = 0  # Track the number of nodes

    def getSize(self):
        """Returns the size of the circular linked list."""
        return self.size

    def insertAtFirst(self, data):
        """Inserts a node at the beginning of the circular linked list."""
        newNode = Node(data)
        print(f"Inserting {data} at the beginning")
        
        if self.head is None:
            # First node case: Point to itself
            self.head = newNode
            newNode.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next  # Traverse to the last node
            
            temp.next = newNode  # Last node points to new node
            newNode.next = self.head  # New node points to head
            self.head = newNode  # Update head to the new node
        
        self.size += 1  # Increment size

    def insertAtLast(self, data):
        """Inserts a node at the end of the circular linked list."""
        newNode = Node(data)
        print(f"Inserting {data} at the end")
        
        if self.head is None:
            self.head = newNode
            newNode.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next  # Traverse to the last node
            
            temp.next = newNode  # Last node points to new node
            newNode.next = self.head  # New node points to head
        
        self.size += 1  # Increment size

    def deleteAtFirst(self):
        """Deletes the first node from the circular linked list."""
        if self.head is None:
            print("List is empty, nothing to delete.")
            return
        
        print(f"Deleting first element: {self.head.data}")
        if self.head.next == self.head:  # Only one node case
            self.head = None
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next  # Traverse to the last node
            
            temp.next = self.head.next  # Last node points to second node
            self.head = self.head.next  # Update head to the second node
        
        self.size -= 1  # Decrement size

    def deleteAtLast(self):
        """Deletes the last node from the circular linked list."""
        if self.head is None:
            print("List is empty, nothing to delete.")
            return
        
        print("Deleting last element")
        if self.head.next == self.head:  # Only one node case
            print(f"Deleted: {self.head.data}")
            self.head = None
        else:
            temp = self.head
            while temp.next.next != self.head:
                temp = temp.next  # Traverse to second last node
            
            print(f"Deleted: {temp.next.data}")
            temp.next = self.head  # Second last node points to head
        
        self.size -= 1  # Decrement size

    def display(self):
        """Displays the circular linked list."""
        if self.head is None:
            print("List is empty")
            return
        
        print("The Circular Linked List is:", end=" ")
        temp = self.head
        for _ in range(self.size):  # Use size to avoid infinite loops
            print(f"{temp.data} -> ", end="")
            temp = temp.next
        print("(back to head)")

# Testing Circular Linked List
cll = CircularLinkedList()
cll.display()
print(f"Size: {cll.getSize()}")

cll.insertAtFirst(1)
cll.insertAtFirst(2)
cll.insertAtLast(3)
cll.insertAtLast(4)
cll.display()
print(f"Size: {cll.getSize()}")

cll.deleteAtFirst()
cll.display()
print(f"Size: {cll.getSize()}")


cll.deleteAtLast()
cll.display()
print(f"Size: {cll.getSize()}")
```

# implementing queue using linkend list


```python
class Node:
    """Represents a node in a linked list."""
    def __init__(self, data):
        self.data = data  # Stores the value of the node
        self.next = None  # Pointer to the next node

class Queue:
    """Implements a queue using a linked list."""
    def __init__(self):
        self.front = None  # Points to the front node
        self.rear = None   # Points to the rear node

    def is_empty(self):
        """Checks if the queue is empty."""
        return self.front is None

    def enqueue(self, data):
        """
        Adds an element to the rear of the queue.

        Args:
            data: The value to be inserted.
        """
        new_node = Node(data)
        if self.rear is None:  # If queue is empty, front and rear are the same
            self.front = self.rear = new_node
            return
        self.rear.next = new_node  # Link new node to the last node
        self.rear = new_node  # Update rear pointer

    def dequeue(self):
        """
        Removes and returns the front element of the queue.

        Returns:
            The value of the dequeued node, or None if the queue is empty.
        """
        if self.is_empty():
            print("Queue is empty!")
            return None
        temp = self.front  # Store the front node
        self.front = self.front.next  # Move front pointer to next node
        if self.front is None:  # If queue becomes empty, update rear
            self.rear = None
        return temp.data

    def peek(self):
        """
        Returns the front element without removing it.

        Returns:
            The front element or None if the queue is empty.
        """
        if self.is_empty():
            print("Queue is empty!")
            return None
        return self.front.data

    def display(self):
        """Displays the elements in the queue."""
        if self.is_empty():
            print("Queue is empty!")
            return
        temp = self.front
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")


# Example usage
if __name__ == "__main__":
    q = Queue()
    q.enqueue(10)
    q.enqueue(20)
    q.enqueue(30)
    q.display()  # Output: 10 -> 20 -> 30 -> None

    print("Dequeued:", q.dequeue())  # Output: Dequeued: 10
    q.display()  # Output: 20 -> 30 -> None

    print("Front element:", q.peek())  # Output: Front element: 20
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None
        self.size=0

    def enqueue(self, data):
        new_node = Node(data)
        print('inserting data:',new_node.data)
        if self.rear is None: # checking if empty
            self.front = self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node
        self.size+=1

    def dequeue(self):
        if self.size==0:
            print("Queue is empty!")
            return None
        temp = self.front
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        self.size-=1
        return temp.data

    def peek(self):
        if self.size==0:
            print("Queue is empty!")
            return None
        return self.front.data

    def display(self):
        if self.size==0:
            print("Queue is empty!")
            return
        temp = self.front
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
        print(self.size)


if __name__ == "__main__":
    q = Queue()
    q.enqueue(10)
    q.enqueue(20)
    q.enqueue(30)
    q.display()

    print("Dequeued:", q.dequeue())
    q.display()

    print("Front element:", q.peek())
```
