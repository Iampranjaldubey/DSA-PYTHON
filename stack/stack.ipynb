```python
# Dynamic size array implementation 
class stack:
    def __init__(self):
        # Initialize an empty list to represent the stack
        self.stack = []

    def push(self, data):
        # Push an element onto the stack
        self.stack.append(data)
        return self.stack

    def pop(self):
        # Remove and return the top element of the stack
        if not self.stack:
            print("Cannot pop, stack is empty")  # Handle case when stack is empty
            return None
        return self.stack.pop()

    def peek(self):
        # Return the top element without removing it
        if not self.stack:
            print('Stack is empty')  # Handle case when stack is empty
            return None
        return self.stack[-1]

    def isEmpty(self):
        # Check if the stack is empty
        return len(self.stack) == 0

    def size(self):
        # Return the number of elements in the stack
        return len(self.stack)

    def show(self):
        # Print the current elements in the stack
        print("Stack:", self.stack)
    
    def reverse(self):
        # Reverse the stack in place
        self.stack.reverse()
        print('Reversed stack:', self.stack)
    
    def sort(self):
        # Sort the stack in ascending order
        self.stack.sort()
        print("Sorted stack:", self.stack)

# Create a stack instance
s = stack()

# Check if the stack is empty
print("Is stack empty?", s.isEmpty())

# Push elements onto the stack
s.push(1)
s.push(9)
s.push(7)
s.push(1)
s.push(5)

# Display stack elements
s.show()

# Pop the top element from the stack
s.pop()

# Display stack elements after popping
s.show()

# Check if the stack is empty
print("Is stack empty?", s.isEmpty())

# Reverse the stack
s.reverse()

# Sort the stack
s.sort()

# Display stack elements after sorting
s.show()
```

```python
# reverse string using stack
def reverse_string(s):
    stack = []  # Create a stack
    for char in s:
        stack.append(char)  # Push all characters onto the stack
    
    reversed_string = ""
    while stack:
        reversed_string += stack.pop()  # Pop characters to form the reversed string
    
    return reversed_string

```

```python
# delete mid element of the stack
def delete_mid(st):
    mid = len(st) // 2  # Find the middle index
    new_stack = []  # Create a new stack to store elements except the middle one
    
    for i in range(len(st)-1,-1,-1):
        if i != mid:  # Skip the middle element
            new_stack.append(st[i])
    new_stack.reverse()
    return (new_stack)

# Test the function
result = delete_mid([1, 2, 3, 4, 5])
print("Stack after deleting middle element:", result)
```

```python
class Browser:
    def __init__(self):
        self.back_stack = []     
        self.forward_stack = []   
        self.current_url = None   

    def visit(self, url: str):

        if self.current_url:
            self.back_stack.append(self.current_url)
        self.current_url = url
        self.forward_stack.clear()

    def back(self):

        if not self.back_stack:
            return self.current_url
        self.forward_stack.append(self.current_url)
        self.current_url = self.back_stack.pop()
        return self.current_url

    def forward(self):

        if not self.forward_stack:
            return self.current_url
        self.back_stack.append(self.current_url)
        self.current_url = self.forward_stack.pop()
        return self.current_url

    def current(self):

        return self.current_url

bh = Browser()
bh.visit("google.com")
bh.visit("youtube.com")
print(bh.back())      # google.com
print(bh.forward())   # youtube.com
bh.visit("github.com")
print(bh.back())
print(bh.back())
print(bh.back())
print(bh.forward()) 
print(bh.forward())
print(bh.forward())

```

# Implementation of stack using different methods 
 

```python
# using array for given capacity 
class Stack:
    def __init__(self,cap):
        self.cap=cap
        self.top=-1
        self.arr=[0]*cap
    
    # push function for adding the element in the stack
    def push(self,data):
        # check if the stack is full 
        if self.top >= self.cap - 1:
            print ("stack overflow")
            return 0
        self.top +=1
        self.arr[self.top]=data
    
    # pop function to delete element from the top(end)
    def pop(self):
        if self.top <0:
            print("stack underflow")
            return 0
        popped=self.arr[self.top]
        self.top -=1
        return popped
    
    # function for peek(Top)
    def peek(self):
        if self.top<0:
            print("stack is empty")
            return 0
        return self.arr[self.top] 
    
    # function to check if the stack is empty
    def is_empty(self):
        return self.top < 0  
    
    # Display function 
    def display(self):
        if self.top < 0:
            print("Stack is empty")
        else:
            print("Stack elements:", self.arr[:self.top + 1])

```

```python
s = Stack(5)
s.push(10)
s.push(20)
s.push(30)
print(s.pop(), "popped from stack")

print("Top element is:", s.peek())
s.display()
print("Elements present in stack:", end=" ")
while not s.is_empty():
    print(s.peek(), end=" ")
    s.pop()
```

```python
# Python program to implement a stack using singly linked list

# Class representing a node in the linked list
class Node:
    def __init__(self, new_data):
        self.data = new_data
        self.next = None

# Class to implement stack using a singly linked list
class Stack:
    def __init__(self):

        # head of the linked list
        self.head = None

    # Function to check if the stack is empty
    def is_empty(self):

        # If head is None, the stack is empty
        return self.head is None 

    # Function to push an element onto the stack
    def push(self, new_data):

        # Create a new node with given data
        new_node = Node(new_data)

        # Check if memory allocation for the new node failed
        if not new_node:
            print("\nStack Overflow")
            return

        # Link the new node to the current top node
        new_node.next = self.head

        # Update the top to the new node
        self.head = new_node

    # Function to remove the top element from the stack
    def pop(self):

        # Check for stack underflow
        if self.is_empty():
            print("\nStack Underflow")
        else:

            # Assign the current top to a temporary variable
            CurrTop = self.head
    
            # Update the top to the next node
            self.head = self.head.next

            # Deallocate the memory of the old top node
            del CurrTop

    # Function to return the top element of the stack
    def peek(self):

        # If stack is not empty, return the top element
        if not self.is_empty():
            return self.head.data
        else:
            print("\nStack is empty")
            return float('-inf')


# Creating a stack
st = Stack()

# Push elements onto the stack
st.push(11)
st.push(22)
st.push(33)
st.push(44)

# Print top element of the stack
print("Top element is", st.peek())

# removing two elemements from the top
print("Removing two elements...");
st.pop()
st.pop()

# Print top element of the stack
print("Top element is", st.peek())
```

```python
# implementing stack using deque
# considering rear as top 
from collections import deque

class Stack:
    def __init__(self):
        self.deque = deque()

    def push(self, item):
        """Add item to the top of the stack (rear of deque)."""
        self.deque.append(item)

    def pop(self):
        """Remove and return the top item from the stack (rear of deque)."""
        if self.is_empty():
            raise IndexError("pop from empty stack")
        return self.deque.pop()

    def peek(self):
        """Return the top item of the stack without removing it."""
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self.deque[-1]

    def is_empty(self):
        """Check whether the stack is empty."""
        return not self.deque

    def size(self):
        """Return the number of elements in the stack."""
        return len(self.deque)
```

```python
# Reinitialize the Stack object to use the correct class
s = Stack()
s.push(10)
s.push(20)
s.push(30)
print(s.peek())  # Output: 30
print(s.pop())   # Output: 30
print(s.pop())   # Output: 20
print(s.is_empty())  # Output: False
print(s.pop())   # Output: 10
print(s.is_empty())  # Output: True
```

```python
# considering front as top 
from collections import deque

class Stack:
    def __init__(self):
        self.dq = deque()

    def push(self, item):
        self.dq.appendleft(item)  # Adds to the front (left)

    def pop(self):
        if not self.dq:
            raise IndexError("pop from empty stack")
        return self.dq.popleft()  # Removes from the front

    def peek(self):
        if not self.dq:
            raise IndexError("peek from empty stack")
        return self.dq[0]         # Front is the top
```
